```python
import pandas as pd
```

```python
df = pd.read_csv("/content/canadian_immegration_data.csv")
```

```python
df
```

```python
df.info()
```

```python
df.set_index('Country', inplace=True)
```

```python
years = list(map(str, range(1980, 2014)))
```

```python
years
```

```python
brazil = df.loc["Brazil", years]
```

```python
brazil
```

```python
brazil_dict = {'year': brazil.index.tolist(), 'immigrants': brazil.values.tolist()}
data_brazil = pd.DataFrame(brazil_dict)
```

```python
data_brazil
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 4))
plt.plot(data_brazil['year'], data_brazil['immigrants'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.yticks([500, 1000, 1500, 2000, 2500, 3000])
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
plt.title('Canadian Immigrants from Brazil from 1980 to 2013')
plt.grid(True)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(data_brazil['year'], data_brazil['immigrants'])
ax.set_title('Canadian Immigrants from Brazil \n from 1980 to 2013')
ax.set_xlabel('Year')
ax.set_ylabel('Number of Immigrants')
ax.grid(True)
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15,5))

axs[0].plot(data_brazil['year'], data_brazil['immigrants'])
axs[0].set_title('Canadian Immigrants from Brazil \n1980 to 2013')
axs[0].set_xlabel('Year')
axs[0].set_ylabel('Number of Immigrants')
axs[0].grid(True)
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))

axs[1].boxplot(data_brazil['immigrants'])
axs[1].set_title('Boxplot of Immigrants from Brazil \n1980 to 2013')
axs[1].set_xlabel('Brazil')
axs[1].set_ylabel('Number of Immigrants')
axs[1].grid(True)

plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Immigration from the four biggest countries of South America to Canada \nfrom 1980 to 2013')

axs[0, 0].plot(df.loc['Brazil', years])
axs[0, 0].set_title('Brazil')

axs[0, 1].plot(df.loc['Colombia', years])
axs[0, 1].set_title('Colombia')

axs[1, 0].plot(df.loc['Argentina', years])
axs[1, 0].set_title('Argentina')

axs[1, 1].plot(df.loc['Peru', years])
axs[1, 1].set_title('Peru')

for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.grid(True)

for ax in axs.flat:
  ax.set_xlabel('Year')
  ax.set_ylabel('Number of Immigrants')

ymin = 0
ymax = 7000
for ax in axs.ravel():
  ax.set_ylim(ymin, ymax)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(data_brazil['year'], data_brazil['immigrants'], lw=3, color='g')
ax.set_title('Canadian Immigrants from Brazil\n from 1980 to 2013', fontsize=18, loc='left')
ax.set_xlabel('Year', fontsize=14)
ax.set_ylabel('Number of Immigrants', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
# ax.grid(linestyle='--')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

fig.savefig('Immigration_brazil_canada.png', transparent=False, dpi=300, bbox_inches='tight')
plt.show()
```

```python
south_america = df.query('Region == "South America"')
```

```python
south_america
```

```python
south_america_sorted = south_america.sort_values('Total', ascending=True)

cores = []
for country in south_america_sorted.index:
  if country == 'Brazil':
    cores.append('green')
  else:
    cores.append('silver')

fig, ax = plt.subplots(figsize=(12,5))
ax.barh(south_america_sorted.index, south_america_sorted['Total'], color=cores)
ax.set_title('South American Countries: Brazil was the fourth country with \nthe most immigrants from 1980 to 2013', loc='left', fontsize=18)
ax.set_xlabel('Number of Immigrants', fontsize=14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

for i, v in enumerate(south_america_sorted['Total']):
  ax.text(v + 20, i, str(v), color='black', fontsize=10, ha='left', va='center')

ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)
ax.tick_params(axis='both', which='both', length=0)

fig.savefig('Immigration_south_america.png', transparent=False, dpi=300, bbox_inches='tight')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h')

ax.set_title('Countries with the most immigrants to Canada \nfrom 1980 to 2013', loc='left', fontsize=16)
ax.set_xlabel('Number of Immigrants', fontsize=14)
ax.set_ylabel('')

plt.show()
```

```python
def generate_graph_palette(palette):
    fig, ax = plt.subplots(figsize=(8, 4))
    ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h', palette=palette, hue=top_10.index, legend=False)
    ax.set_title('Countries with the most immigrants to Canada \nfrom 1980 to 2013', loc='left', fontsize=18)
    ax.set_xlabel('Number of Immigrants', fontsize=14)
    ax.set_ylabel('')
    plt.show()
```

```python
import plotly.graph_objects as go
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(x=[data_brazil['year'].iloc[0]], y=[data_brazil['immigrants'].iloc[0]],
                         mode='lines',
                         name='Immigrants',
                         line=dict(color='green', width=3)))

fig.update_layout(title=dict(
                            text='<b>Immigration from Brazil to Canada from 1980 to 2013</b>',
                            x=0.12,
                            xanchor='left',
                            font=dict(size=20)
                            ),
                  xaxis=dict(range=[1980, 2013], autorange=False, title='<b>Year</b>'),
                  yaxis=dict(range=[0, 3000], autorange=False, title='<b>Number of Immigrants</b>'),
                  updatemenus=[dict(
                      type='buttons',
                      showactive=False,
                      buttons=[dict(
                          label='Play',
                          method='animate',
                          args=[None, {'frame': {'duration': 100, 'redraw': True}, 'fromcurrent': True}]
                      )]
                  )],
                  width=1000,
                  height=500
                  )

frames = [go.Frame(data=[go.Scatter(x=data_brazil['year'].iloc[:i+1], y=data_brazil['immigrants'])]) for i in range(len(data_brazil))]
fig.frames = frames

fig.show()
```
